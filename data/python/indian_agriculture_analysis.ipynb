```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/cleaned_datafile (1).csv')
# basic data inspection
df.head(10)
df.shape
df.info()
df.describe()

```

```python
#Check Missing Values
df.isnull().sum()
# there is no missing values is found
```

```python
corr=df.select_dtypes(include='number').corr()
print(corr)
#heatmap using Matplotlib
plt.figure(figsize=(8,6))
plt.imshow(corr)
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Heatmap")
plt.show()
```

 insights of correlation:

1. Yield and Yield Efficiency show a very strong positive correlation (0.97), indicating that crops with higher yields tend to be more efficient.

2. Total Cultivation Cost and Yield have a strong positive correlation (0.87), suggesting that higher cultivation investments are generally associated with higher productivity.

3. Direct Cultivation Cost and Total Cultivation Cost are highly correlated (0.98), showing that direct cultivation expenses make up a major portion of total cultivation costs.

4. Extra Cultivation Cost and Total Cultivation Cost also exhibit a strong positive correlation (0.93), indicating that additional cultivation expenses significantly contribute to overall costs.

5. Extra Cultivation Cost and Yield have a strong positive correlation (0.78), suggesting that additional investments may help improve crop output.

6. Cost of Production per Quintal has a moderate negative correlation with Yield (-0.49), meaning that higher-yielding crops tend to have lower production costs per unit.

7. Cost of Production per Quintal and Yield Efficiency have a moderate negative correlation (-0.56), indicating that higher production costs are associated with lower efficiency.

8. Overall, cultivation costs show positive relationships with yield and efficiency, while production cost per quintal shows negative relationships with both.
